```python
import numpy as np
import matplotlib as mpl
import pandas as pd
from matplotlib import pyplot as plt

# we need the following line to indicate that the plots should be shown inline with the Jupyter notebook.
%matplotlib inline

# Survey class data
answers = pd.read_csv('survey.csv', sep=',', low_memory=False, encoding = 'ISO-8859-1')

answers.head()
```

```python
x_age = answers['Your age (in years)'].tolist()
y_height = answers['Your height (in cm)'].tolist()

plt.plot(x_age, y_height, 'ro')
```
